# On Git & the GitHub and GitLab Platforms:

## About Git

Git is an Open Source Distributed Version Control System for tracking changes in source code or any other set of software files.

- **Control System:** Git is a content tracker.

- **Version Control System:** Code is constantly changing. Many developers can add code in parallel. Git keeps history of what changes have been implemented.

- **Branching/Forks:** Git provides features like branches and merges.

- **Distributed Version Control System:** Git managers, like GitHub, have a remote repository stored in a server and a local repository which is stored in the computer of each developer.

## About GitHub

- GitHub is a web-based platform that provides hosting for software development and version control using Git.

- GitHub is a company that provides hosting for software development version control using Git.

- It is an ideal platform to share code with colleagues.

- The (rolling-course) CP1 GitHub Repository serves this purpose: https://github.com/wbandabarragan/computational-physics-1

- You have afew options:

  - **Watch:** this helps you view and keep track of all changes.

  - **Fork:** this creates a mirror version of the repository in your own GitHub accounts.

## Using Git (CLI option):

Download the CP1 material from the remote repository to your own laptop:

1. Go to https://github.com/wbandabarragan/computational-physics-1

2. Open a terminal, navigate to the desired working directory and create a sub-directory. Then, inside the sub-directory:

```
git clone https://github.com/wbandabarragan/computational-physics-1.git
```

Now you have all the current material in your own computer.

3. Updating the repository requires:

```
git pull
```

4. Adding and modifying the local repository can be done in your own laptops/PCs. Once you have finished the changes, you can update the remote repository with:

```
git add .
```

5. That will stash all changes, but you still need to add a comments before commiting the changes:

```
git commit -m “Update repo.”
```

6. Pushing the changes to the remote repository:

```
git push
```

All these commands should work with your own repository or in any repository in which you are contributor

#### Note: Please note that you will need to set up your own keys in the secutiry settings of your GitHub accounts to be able to commit changes to your repositories.

### Using git: branches

A few additional commands that are helpful include:

1. `git status`: this is to check any updates in the repository.

```
git status

On branch main
Your branch is up to date with 'origin/main'.

Untracked files:
  (use "git add <file>..." to include in what will be committed)
	.ipynb_checkpoints/
	echo.txt
	echo.txt.asc
	first-notebook.ipynb

nothing added to commit but untracked files present (use "git add" to track)
```

2.`git branch`: this is to check how many branches there are in the github repository:

```
git branch
* main
```

3. `git branch test`: this is to create a new branch

```
git branch test
git branch
* main
test
```

4. `git checkout test`: this is to inspect the new branch:

```
git checkout test
Switched to branch test

git branch
main
* test
```

5. To merge main branch with test branch:

```
git checkout main
git merge test
```

## Managing Software with GUI-based software:

There are also Git managers that have graphical interfaces. These GUI-based applications allow you to do the same we were doing with the CLI.

- **GitKraken:**

- Gitkraken is a popular graphical user interface (GUI) for Git.

- Gitkraken is a cross-platform desktop client for Linux, macOS, and Windows that provides a visual interface to interact with Git repositories.

- It's designed to simplify complex Git commands and workflows, making version control easier, particularly for both beginners.

- Many developers use both, the Git CLI for quick, routine tasks and GitKraken for more complex operations, visualisation, and team collaboration.

### On GitLab (an alternative to GitHub):


- GitLab is a comprehensive, web-based platform that provides similar capabilities to GitHub for managing repositories.

- Github aims to provide a **single application for the entire software development lifecycle**, from project planning and source code management to continuous integration (CI), continuous delivery (CD), and security.

- Unlike GitHub, which relies on third-party integrations for many of these features, GitLab builds them directly into the platform, creating a more cohesive and streamlined workflow.

- This "all-in-one" approach, along with its strong support for both cloud-based and self-hosted deployments, makes GitLab particularly popular with teams developing software from scratch and who want to centralise their tools and automate their processes from a single interface.

- **Both GitHub and GitLab are widely used in physics.**

#### Reference: https://about.gitlab.com
